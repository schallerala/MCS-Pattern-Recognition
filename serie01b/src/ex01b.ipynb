```python
from loading_data import load_dataframe

import matplotlib.pyplot as plt
%matplotlib inline

Y, X = load_dataframe('../train.csv', nrows=200)
classes = [i for i in range(10)]
num_classes = len(classes)
samples = 6

# Print a matrix of greyscale to have an idea of how the feature look like
for y in classes:
    samples_indexes = Y[Y == y].sample(samples).index
    for i, sample_index in enumerate(samples_indexes):
        plt_index = i * num_classes + y + 1
        plt.subplot(samples, num_classes, plt_index)
        plt.imshow(X.iloc[sample_index].to_numpy().reshape((28, 28)))
        plt.axis("off")
        if i == 0:
            plt.title(str(y))

plt.show()

```

## Printing a few instances by cluster

_The titles of the plots are the labels of the clusters, and not the recognized digit_

```python
from KMeans import KMeans

model = KMeans(k_clusters=10, random_seed=0).fit(X)

for y in classes:
    samples_indexes = X[model.labels == y].sample(samples).index
    for i, sample_index in enumerate(samples_indexes):
        plt_index = i * num_classes + y + 1
        plt.subplot(samples, num_classes, plt_index)
        plt.imshow(X.iloc[sample_index].to_numpy().reshape((28, 28)))
        plt.axis("off")
        if i == 0:
            plt.title(str(y))

plt.show()
```

## Model validation

```python
print(f"C-Index: {model.c_index_score()}")
print(f"Dunn-Index: {model.dunn_index_score()}")
```
